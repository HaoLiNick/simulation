### Feature Engineering

```python
class ExponentialSaturation(BaseEstimator, TransformerMixin):
    def __init__(self, saturation=1):
        self.saturation = saturation
        
    def fit(self, X, y=None):
        self.feature_names_in_ = _get_feature_names(X)
        X = check_array(X)
        self._check_n_features(X, reset=True)
        return self

    def transform(self, X):
        check_is_fitted(self)
        X = check_array(X)
        self._check_n_features(X, reset=False)
        temp = 1 - np.exp(-self.saturation*X)
        return temp
    
    def get_feature_names_out(self, input_features=None):
        return _check_feature_names_in(self, input_features)
```

```python
class ExponentialCarryover(BaseEstimator, TransformerMixin):
    def __init__(self, strength: float = 1, length: int = 7):
        self.strength = strength
        self.length = length
        
    def fit(self, X, y=None):
        self.feature_names_in_ = _get_feature_names(X)
        X = check_array(X)
        self._check_n_features(X, reset=True)
        self.sliding_window_ = (
            self.strength ** np.arange(self.length + 1)
        ).reshape(-1, 1)
        return self

    def transform(self, X: np.ndarray):
        check_is_fitted(self)
        X = check_array(X)
        self._check_n_features(X, reset=False)
        convolution = convolve2d(X, self.sliding_window_)
        if self.length > 0:
            convolution = convolution[: -self.length]
        return convolution
    
    def get_feature_names_out(self, input_features=None):
        return _check_feature_names_in(self, input_features)
```

```python
class SigmoidSaturation(BaseEstimator, TransformerMixin):
    def __init__(self, saturation: float = 1e-1, half_life: float = 0.5):
        self.saturation = saturation
        self.half_life = half_life
        
    def fit(self, X, y=None):
        self.feature_names_in_ = _get_feature_names(X)
        X = check_array(X)
        self._check_n_features(X, reset=True)
        return self

    def transform(self, X):
        check_is_fitted(self)
        X = check_array(X)
        self._check_n_features(X, reset=False)
        return 1 / (1 + np.exp(-self.saturation * (X - X.mean())))
    
    def get_feature_names_out(self, input_features=None):
        return _check_feature_names_in(self, input_features)
```

```python
def negative_mean_absolute_error_percentage(y_true: np.ndarray, y_pred: np.ndarray) -> float: 
    """
    Computes negative mean absolute error percentage. This is the mean absolute error normalized by the sum of the actuals. The metric
    is negative to allow for max optimizing.
    
    Parameters:
    -----------
    y_true: np.ndarray
        Array of actual/observed values
    y_pred: np.ndarray
        Array of predicted values
        
    Returns:
    --------
    negative_mean_absolute_error_percentage: float
        negative mean absolute error normalized by the sum of the actuals
    """
    epsilon = np.finfo(np.float64).eps
    
    total_absolute_residuals = np.sum(np.abs(y_pred - y_true))
    total_actuals = np.sum(y_true)

    return -1 * total_absolute_residuals / (total_actuals + epsilon)


neg_maep = make_scorer(negative_mean_absolute_error_percentage)
```

```python
def negative_normalized_root_mean_squared_error(y_true: np.ndarray, y_pred: np.ndarray) -> float: 
    """
    Computes negative normalized root mean squared error. This is the root mean absolute error normalized by the range of the actuals. The metric
    is negative to allow for max optimizing.
    
    Parameters:
    -----------
    y_true: np.ndarray
        Array of actual/observed values
    y_pred: np.ndarray
        Array of predicted values
        
    Returns:
    --------
    negative_normalized_root_mean_squared_error: float
        negative normalized mean squared error normalized by the sum of the actuals
    """
    epsilon = np.finfo(np.float64).eps
    
    mean_squared_error = np.sum((y_pred - y_true)**2)
    root_mean_squared_error = np.sqrt(mean_squared_error)
    
    y_range = y_true.max() - y_true.min()
    negative_normalized_root_mean_squared_error = root_mean_squared_error / (y_range + epsilon)

    return -1 * negative_normalized_root_mean_squared_error


neg_nrmse = make_scorer(negative_normalized_root_mean_squared_error)
```
